```python
import os
import random
import subprocess

def findFile(source, i):
    for file in os.listdir(source):
        nn = str(i) + "F"
        if os.path.splitext(file)[0] == nn:
            return os.path.join(source, file)
    return None

def randomFile(source, i):
    while True:
        rand_number = random.randint(1, 200)
        if rand_number != i:
            tar = findFile(source, rand_number)
            if tar is not None:
                print(f"Target File: {tar}")
                return tar, rand_number

#source = input("Enter the source directory: ")
source = "/content/FaceDancer/inside/INSIDE/"

for i in range(200):
    src = findFile(source, i)
    if src is None:
        continue

    print(f"Source File: {src}")
    tar, rand_number = randomFile(source, i)

    # Command as a list
    command = [
        "python", "test_image_swap_multi.py",
        "--facedancer_path", "./model_zoo/FaceDancer_config_c_HQ.h5",
        "--img_path", tar,
        "--swap_source", src,
        "--img_output", f"./results/inside/front_c_HQ/{os.path.basename(src)}_{os.path.basename(tar)}.jpg"
    ]

    subprocess.run(command)
```
